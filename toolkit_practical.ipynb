```python
"""
QSN-4-How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
ANS-
"""
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.random.seed(0)
data = np.random.rand(100, 5)
df = pd.DataFrame(data, columns=['task1', 'task2', 'task3', 'task4', 'task5'])
corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('answer of question 4')
plt.show()
```

```python
"""
QSN-6-Create a DataFrame and add a new column based on an existing column.
ANS-
"""
import pandas as pd

data = {'Name': ['ajay', 'bijay', 'sanjay', 'jay'],'Age': [28, 24, 35, 32],'Score': [85, 90, 78, 92]}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

df['Grade'] = pd.cut(df['Score'], bins=[0, 70, 85, 100], labels=['F', 'P', 'A'], include_lowest=True)

print(df)
```

```python
"""
QSN-7-Write a program to perform element-wise multiplication of two NumPy arrays.
ANS-
"""
import numpy as np
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([6, 7, 8, 9, 10])
print("Array 1:")
print(array1)
print("\nArray 2:")
print(array2)

result = np.multiply(array1, array2)

print(result)

```

```python
"""
QSN-8-Create a line plot with multiple lines using Matplotlib.
ANS-
"""
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(2x)')

plt.title('Line Plot with Multiple Lines')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
```

```python
"""
QSN-9-Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
ANS-
"""
import pandas as pd
import numpy as np

np.random.seed(0)
data = {
    'Name': ['ajay', 'bijay', 'sanjay', 'jay', 'ranvijay', 'sujoy'],
    'Age': [28, 24, 35, 32, 40, 38],
    'Score': np.random.randint(0, 100, 6)
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
threshold = 70
filtered_df = df[df['Score'] > threshold]
print("\nFiltered DataFrame (Score > {}):".format(threshold))
print(filtered_df)

```

```python
"""
QSN-10-Create a histogram using Seaborn to visualize a distribution.
ANS-
"""
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
data = np.random.randn(1000)
plt.figure(figsize=(8, 6))
sns.histplot(data, bins=30, kde=True)
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
"""
QSN-11-Perform matrix multiplication using NumPy.
ANS-
"""
import numpy as np
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])
print("Matrix A:")
print(matrix_a)
print("\nMatrix B:")
print(matrix_b)
result = np.matmul(matrix_a, matrix_b)
print(result)
```
